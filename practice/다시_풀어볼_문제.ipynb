### # Binary Search ì´ì§„ íƒìƒ‰

ì¤‘ê°„ê°’ì„ í™•ì¸í•˜ê³ , ë‚´ê°€ ì°¾ëŠ” íƒ€ê²Ÿì´ ë” í¬ë©´?

ì‘ì€ ì™¼ìª½ì€ ë‹¤ ë²„ë¦¬ê³  ë‚˜ë¨¸ì§€ ì ˆë°˜ì—ì„œ ì¤‘ê°„ í™•ì¸



```python
numbers = range(1, 51)
target = 23
result = -1

left = 0
right = len(numbers) -1

while left <= right: # ì›ë˜ëŠ” Trueë¼ê³  ì¼ëŠ”ë° ë§Œì•½ ì°¾ëŠ” ê°’ì´ ì—†ìœ¼ë©´ í¬ë¡œìŠ¤ê°€ ë˜ì–´ë²„ë ¤ì„œ ì´ë ‡ê²Œ ëª…ëª…í•´ì¤˜ì•¼ë¨
    mid = (left+right) // 2
    print("=" * 10)
    print(left, right, mid)
    if numbers[mid] == target:
        result = mid
        break
    elif numbers[mid] > target: # ê°€ìš´ë°ê°€ íƒ€ê²Ÿë³´ë‹¤ í¬ë‹¤
        right = mid - 1
    else: # ì‘ë‹¤
        left = mid + 1
print(result)
```
