### 문제
1번에서 나는 if target in s: 라고 했는데.. 별로 안 좋은 코드였다

분석을 할 때 "정확히 이 단어인 것만 골라내!" 할 때는 ==를 쓰고, "이 단어가 포함된 문장을 다 찾아내!" 할 때는 in을 씁니다.

예: if user_id == "admin": (관리자 아이디와 똑같은지 확인)

예: if "비속어" in comment: (댓글에 나쁜 말이 포함되어 있는지 확인)

```python
# 1. s 개수 찾기
st = "starbucks"
target = "s"
cnt = 0

for s in st:
    if target == s:
        cnt += 1
print(cnt)
```

```python
# 2. star 찾기 <<이 부분 혼자 작성해보기!!!!!!!!!!>>
st = "starbucks"
target = "star"
cnt = 0

for i in range(len(st)):
    print(st[i : i+len(target)]) # st[0:4] 4는 타겟의 길이

for i in range(len(st) - len(target)+1): # 뒤에 cks, ks, s 버리기
    if st[i : i+len(target)] == target:
        cnt += 1
print(cnt)


# 다른 방법
st = "starbucks"
target = "star"
target_len = len(target)
cnt = 0

for i in range(len(st)):
    if st[i : i + target_len] == target:
        cnt += 1
print(cnt)
```

```python
# 3. 약수 찾기
target = 10
result = []
# 1, 2, 5, 10은 타겟을 나누었을 때 0이 되는
for i in range(1, target+1):
    if target % i == 0:
        result.append(i)
print(result)


# 4. 효율적으로 약수 찾기 (계산량을 절반으로 줄이기)
target = 2048
result = [1]
for i in range(2, target // 2 + 1): # 1은 미리 추가해두고, 절반까지만 돌리기 # 몫으로
    if target % i == 0:
        result.append(i)
result.append(target)
print(result)
```
