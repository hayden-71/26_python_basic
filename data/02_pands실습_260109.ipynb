```python
import pandas as pd

# 1. 데이터 불러오고 훑어보기

movie_df = pd.read_csv('korea_movie_best.csv')
print(movie_df.head())
print("-"*20)
print(movie_df.info())
```

```python
# [미션 2] 데이터 선택하기 (Indexing)
# 영화 제목(영화명)을 인덱스로 지정(set_index)하여 새로운 변수 df_index에 저장하세요.
df_index = movie_df.set_index('영화명')
print(df_index)

# loc을 사용하여 영화 '기생충'의 '평점'을 찾아 출력하세요.
print('\n기생충 평점', df_index.loc['기생충', '평점'])
```

```python
# [미션 3] 조건으로 필터링하기 (Boolean Indexing)
# 관객수가 1400만 명을 넘는 초대박 영화만 골라내서 hit_movie 변수에 저장하세요.
hit_movie = df_index[df_index['관객수'] > 1400]
print(hit_movie)

# 장르가 '코미디'인 영화만 골라내서 출력하세요.
comedy = df_index[df_index['장르'] == '코미디']
print('\n코미디', comedy)
```

```python
# [미션 4] 데이터 정렬 및 파생변수 (Sorting & Calculation)
# 전체 영화를 평점이 높은 순서대로(내림차순) 정렬해서 1위부터 3위까지만 출력하세요.
score_top3 = df_index.sort_values(by='평점', ascending=False)[:3] #.head(3) 써도 됨
print(score_top3)

# 영화 티켓 가격이 10,000원이라고 가정합니다. 관객수 * 10,000 을 계산하여 매출액이라는 새로운 컬럼을 만드세요
df_index['매출액'] = df_index['관객수'] * 10000
df_index[['관객수', '평점']].head()
```

### 드러운 데이터 처리하기

```python
import numpy as np

dirty_data = {
    '날짜': ['2026-01-01', '2026-01-02', '2026-01-02', '2026-01-03', '2026-01-04'],
    '상품명': ['아메리카노', '카페라떼', '카페라떼', '바닐라라떼', np.nan],
    '가격': ['4,000', '4,500', '4,500', '5000', '5,500'],
    '판매수량': [50, 30, 30, np.nan, 15]
}

df = pd.DataFrame(dirty_data)
print("--- 원본 데이터 ---")
print(df)
print("\n--- 정보 확인 ---")
print(df.info())
```

```python
print("--- 1. 결측치 확인 ---")
print(df.isnull().sum()) # col별 빈 칸 개수 세기
```

```python
# 1. 결측치 있는 행 제거 (원본 보호를 위해 새로운 변수에!!)
df_dropped = df.dropna()
print("[삭제 후]")
print(df_dropped)
```

```python
# 2.결측치 채우기 (판매수량이 비어있으면 0으로 채움)
# 원본 df에 바로 적용하려면 inplace=True를 쓰거나 재할당
df['판매수량'] = df['판매수량'].fillna(0)
print("[0으로 채운 후]")
print(df)
```

```python
# 상품명은 채울 수 없으니 '알수없음'으로 채우기
df['상품명'] = df['상품명'].fillna('IDK??')
print(df)
```

```python
print("--- 2. 중복 제거 전 ---")
print(len(df))

# 중복된 행 제거 (keep='first'는 첫번째만 남기고 뒤에 나온 건 지움)
df = df.drop_duplicates()

print("\n--- 중복 제거 후 ---")
print(df)
```

```python
print("--- 3. 타입 변환 ---")

# 1. 쉼표 없애기
df['가격'] = df['가격'].str.replace(',', '')

# 2. int로 변환
df['가격'] = df['가격'].astype(int)

print("\n 변환 후 정보")
print(df.info()) # 가격이 int로 바뀐 것 확인
print("\n평균 가격", df['가격'].mean())
```

### user data

```python
user_data = {
    '가입일': ['2023.01.05', '2023.12.25', '2024.05.05'],
    '이메일': ['kim@naver.com', 'lee@gmail.com', 'park@daum.net'],
    '나이': [25, 35, 19]
}

df_user = pd.DataFrame(user_data)
print(df_user)
```

```python
print("--- 1. 아이디만 추출 (split) ---")
# @ 기준으로 쪼개고, 그 중 첫번째 가져오기
df_user['아이디'] = df_user['이메일'].str.split('@').str[0]
display(df_user)

print("--- 2. 특정 글자 포함 여부 (contains) ---")
# 네이버 사용자만 찾기 (엑셀의 필터)
naver_user = df_user[df_user['이메일'].str.contains('naver')]
print(naver_user)
```

```python
print("--- 3. 날짜 변환 및 정보 추출 ---")

# 1. 글자 -> 날짜로 변환 (형식 지정이 필요할 수 있음)
df_user['가입일'] = pd.to_datetime(df_user['가입일'])
print(df_user['가입일'].info())

# 2. 연, 월, 요일 추출 (.dt 사용)
df_user['가입연도'] = df_user['가입일'].dt.year
df_user['가입월'] = df_user['가입일'].dt.month
df_user['가입요일'] = df_user['가입일'].dt.day_name() # 영어 요일

print(df_user)
```

```python
print("--- 4. apply 함수 활용 ---")

#  나이에 따라 '청년'과 '중년'을 구분하는 함수 만들기
def get_age_group(age):
    if age < 30:
        return '청년'
    else:
        return '중년'
    
# 이 함수를 '나이' 컬럼의 모든 데이터에 적용
df_user['연령대'] = df_user['나이'].apply(get_age_group)

display(df_user)
```
