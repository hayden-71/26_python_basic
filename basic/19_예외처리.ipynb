### 예외처리

```python
# 예외처리란? 프로그래밍 중간에 발생하는 오류를 예상하여 회피하는 방법
# try:
#   오류가 발생할 수 있는 실행문
# except:
#   오류가 발생했을 때 처리할 구문
```

```python
try:
    num1 = 4
    num2 = 0
    print(num1 / num2)
except:
    print("오류가 발생했어요")
```

```python
# 예상 가능한 오류는 별도로 처리가 가능
# "구체적인 에러일수록 위에, 포괄적인 에러일수록 아래에" 적어야 해요.
try:
    4 / 0
    # 1 + 'A'
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다")
except:
    print("이건 무슨일이죠?")
```

```python
# 오류가 발생했을 때와 그렇지 않을 때 처리
try:
    age = int(input("나이를 입력하세요: "))
except:
    print("문제가 발생했습니다.")
else:
    print(f"{age}살 입니다.")
```

```python
# 오류가 발생해도 꼭 처리해야 할 일이 있을 때
price = 1000
try:
    tip = int(input("팁을 입력하세요: "))
    price += tip
except:
    print("문제가 발생했습니다.")
finally:
    print(f"{price:,}원 결제합니다.")
```

```python
# 오류 일부러 발생시키기
raise NotImplementedError
```

```python
# 나만의 에러 만들기

class MyError(Exception):
    def __str__(self):
        return "내가 만든 에러"

try:
    raise MyError
except MyError as e:
    print(e)
except:
    print("???")
```
