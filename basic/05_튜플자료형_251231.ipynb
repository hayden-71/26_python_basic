### 튜플

```python
# 리스트와 거이 동일
# [] 대신 ()
# 내부 요소를 변경할 수 없다
```

```python
t1 = ()
t2 = (1, ) # 하나 들어있을 때는 , 찍어줘야 튜플로 인지
t3 = (1, 2, 3)
t4 = 1, 2, 3 # 여러개면 괄호 없어도 자동으로 튜플로 인식
t5 = (1, 2, (3, 4)) # 튜플 안에 튜플

# 문자열, 딕셔너리, 리스트도 들어갈 수 있음!
```

```python
# index 접근
t = (1, 2, 3)
print(t[0])
print(t[1:])
```

```python
t1 = (1, 2, 3)
t2 = (4, 5, 6)
t3 = t1 + t2
print(t3)
```

```python
t = (0, ) * 10
print(t)
```

```python
# 요소를 변경하려고 하면...
t = (1, 2, 3)
t[0] = 4
# del t[0] 삭제도 마찬가지

# 선언하는 순간 끝이기 때문에 변경하면 안 되는 데이터 다루기 좋음!
```

```python
# 변수 선언
(a, b) = (1, 2)
print(a, b)
```

```python
# 변수 선언
a, b = 1, 2
c, d = (3, 4)
print(a, b, c, d)
```
