### 형변환 (casting)

```python
# 다음 결과 값을 확인하세요

n1 = "3"
n2 = "5"
print(n1 + n2)

n1 = 3
n2 = "5"
print(n1 + n2)
```

```python
# 데이터 타입 간의 변환, 형변환
# 바꾸고자 하는 타입(값)
# ex: int(str)
# ex: str(int)

n1 = "3"
n2 = "5"
print(int(n1) + int(n2))
n3 = "^^"
print(int(n3))
```

```python
# 데이터 타입 간의 변환, 형변환
# 바꾸고자 하는 타입(값)
# ex: int(str)
# ex: str(int)

n1 = 3
n2 = 5
print(str(n1) + str(n2))
print("%d"%n1 + "%d"%n2)
```
