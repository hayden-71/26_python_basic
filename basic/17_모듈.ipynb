### 모듈

```python
# 모듈 : 함수나 변수 또는 클래스를 모아 놓은 파이썬 파일
# 다른 누군가가 미리 짜둔 파이썬 코드라 생각하면 됨
# 다른 파이썬 파일에서 불러와 사용할 수도 있음
```

```python
# 쥬피터 말고 mod1.py에 작성
def add(a, b):
    return a + b


def sub(a, b):
    return a - b

# __name__ : python 내 특수한 변수로써 실행 주체가 누구인지 적혀있는 변수
# 아래 코드의 의미: 해당 파일이 직접 실행(main)된건지 판단
# python mod1.py로 실행해보기
if __name__ == "__main__":
    print(add(5, 6))
```

```python
# 동일 디렉토리 내 mod2.py 작성
# 모듈 불러오기
# import : 이미 만들어 놓은 파이썬 모듈을 불러오는 명령어

import mod1

print(f"3 + 4 = {mod1.add(3, 4)}")
print(f"6 - 2 = {mod1.sub(6, 2)}")
```

```python
# 만약 모듈 이름을 함수 앞에 쓰는게 불편하다면,
# from을 이용해서 다음과 같이 작성하면 함수만 사용 가능
from mod1 import add, sub

print(f"3 + 4 = {add(3, 4)}")
print(f"6 - 2 = {sub(6, 2)}")
```

```python
# datetime from을 사용하면
# datetime.datetime.now() 대신 datetime.now() 가능

# import datetime
# from datetime import *
from datetime import datetime

now = datetime.now()
print(now)
```
