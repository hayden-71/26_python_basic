### 변수

```python
# 데이터를 담아두는 그릇 = 변수
a = 10  # a라는 그릇에 10이라는 데이터를 담아두겠다.
b = 3   # b라는 그릇에 3이라는 데이터를 담아두겠다.
a + b   # a 그릇에 담긴 데이터와 b 그릇에 담긴 데이터를 더하라
```

```python
c = a + b  # a 그릇과 b 그릇에 담긴 데이터를 더한 결과를 c 그릇에 담아두겠다.
c
```
