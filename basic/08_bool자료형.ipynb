### bool 자료형

```python
# bool 자료형은 참(True)과 거짓(False)를 나타내는 자료형
# 두개의 값 True, False
```

```python
t = True
f = False
```

```python
1 == 1 # 1 과 1이 같은가?
```

```python
1 != 2 # 1 과 2가 다른가?
```

```python
10 > 1 # 10이 1보다 큰가?
```

```python
10 >= 1 # 10이 1 이상 인가?
```

```python
10 < 1 # 10이 1보다 작은가?
```

```python
10 <= 1 # 10이 1 이하 인가?
```

```python
# 자료형의 값으로도 True, False를 구분할 수 있습니다
# 문자열
"hello"     # True
""          # False
# List
[1, 2, 3]   # True
[]          # False
# Tuple
(1, 2, 3)   # True
()          # False
# Dict
{'a': 1}    # True
{}          # False
# 상수
123         # True
0           # False
# None
None        # False
```

```python
print(bool("hello"))
print(bool(""))
```
