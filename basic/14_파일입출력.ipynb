### 파일입출력

```python
# 파일_객체 = open(파일_이름(경로), 파일_모드)

# <파일 모드 종류>
# r : 읽기 모드 - 파일을 읽을 때 사용
# w : 쓰기 모드 - 파일을 쓸 때 사용
# a : 추가 모드 - 파일의 마지막에 새로운 내용을 추가할 때 사용
```

```python
f = open("new.txt", "w", encoding="utf-8")
f.close()
```

```python
# new.txt 파일 "쓰기 모드"로 열기
# 파일이 없으면 생성
# f.close() - 파일 닫기 open과 항상 쌍으로 하기
```

```python
# 없는 파일을 "읽기 모드"로 열면 오류 발생
f = open("hello.txt", "r", encoding="utf-8")
f.close()
```

```python
# 시스템 관련 라이브러리
import os

# 해당 파일, 경로가 존재하는지 확인
if os.path.exists("hello.txt"):
    print("파일 있어요!")
    f = open("hello.txt", "r", encoding="utf-8")
    f.close()
else:
    print("파일 없어요!")
```

```python
# 파일 데이터 쓰기
# 쓰기모드는 새로 파일을 씁니다 => 기존 데이터 삭제
f = open("hello.txt", "w", encoding="utf-8")

for i in range(1, 6):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)

dataset = []
for i in range(6, 10):
    dataset.append(f"{i}번째 줄입니다.\n")
f.writelines(dataset)
f.close()
```

```python
# with open 으로 코드를 짜면 f.close()를 안써도 됨(좀 더 간편하게 코딩 가능)
with open("hello.txt", "w", encoding="utf-8") as f:
    for i in range(1, 5):
        data = f"{i}번째 줄입니다.\n"
        f.write(data)
```

```python
# 파일 읽기 - 첫번째 줄 읽기기
f = open("hello.txt", "r", encoding="utf-8")
line = f.readline()
print(line)
f.close()
```

```python
# 모든 내용 읽기
f = open("hello.txt", "r", encoding="utf-8")
while True:
    line = f.readline()
    if not line:
        break
    print(line)
f.close()
```

```python
# 모든 내용 읽기 with 버전
with open("hello.txt", "r", encoding="utf-8") as f:
    for line in f:
        print(line)
```

```python
# 파일 수정하기
with open("hello.txt", "a", encoding="utf-8") as f:
    f.write("몇번째 줄이더라?\n")
```

```python
# 실제 파일 읽어서 처리해보기기
# https://www.datablist.com/learn/csv/download-sample-csv-files
with open("./organizations-100.csv", "r", encoding="utf-8") as f:
    for line in f:
        words = line.split(",")
        if words[4] == "Korea":
            print(words)
```
