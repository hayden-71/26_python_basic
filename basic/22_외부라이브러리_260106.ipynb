### 외부 라이브러리

```python
# 파이썬 표준 라이브러리 외 외부에서 추가하는 라이브러리
# pip를 활용하여 외부 패키지 설치 가능

# pip install [PACKAGE]
# 특정 버전 설치
# pip install [PACKAGE]==VERSION
# 업그레이드
# pip install --upgrade [PACKAGE]
# 삭제
# pip uninstall [PACKAGE]
# 설치 확인
# pip list
```

```python
# 가상환경

# 프로젝트마다 필요한 라이브러리들이 다름
# 동일한 라이브러리도 특별한 버전이 필요할 수 있음
# 이를 관리하기 위해 가상환경을 만들고 해당 공간에 라이브러리 설치

# 가상환경 만들기
# python -m venv venv >> python module 중에 venv를 실행, venv라는 이름으로 만들어달라

# window에서 가상환경 실행
# .\venv\Scripts\activate
# powershell 권한문제 : Set-ExecutionPolicy -Scope Process -ExecutionPolicy Bypass 후 다시 시도

# mac에서 가상환경 실행
# source ./venv/bin/activate

# 실행이 정상적으로 되면, 터미널 앞에 (venv) 표기

# 종료할 때는 deactivate
```

```python
# faker: 가짜 데이터를 만들어주는 라이브러리
# https://pypi.org/project/Faker/

from faker import Faker
fake = Faker()

print(fake.name())
print(fake.address())
```

```python
# art: 시각적으로 글씨 써주기
# https://pypi.org/project/art/
from art import tprint

tprint("PYTHON", font="block")
tprint("HELLO", font="random")
```

```python
# 텍스트를 음성으로
# https://pypi.org/project/gTTS/
from gtts import gTTS

text = "안녕하세요. 파이썬으로 목소리를 만들고 있어요."

# lang='ko'는 한국어
tts = gTTS(text=text, lang='ko')
tts.save("voice.mp3")

print("voice.mp3 파일이 생성되었습니다. 실행해보세요!")
```
