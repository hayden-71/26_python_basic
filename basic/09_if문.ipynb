### if 문

```python
# 프로그래밍 도중 흐름을 제어할 때 활용
# 만약 유저가 신규 유저라면?
# 유저의 결제 수단이 카드라면, 그렇지 않고 현금이라면?
# if [조건]:
#   조건이 참 일 때 수행할 내용
# else:
#   조건이 거짓일 때 수행할 내용
```

```python
num1 = 10
num2 = 1

if num1 > num2:
    print("num1이 num2보다 큽니당")
```

```python
num1 = 10
num2 = 1
if num1 > num2:
    print(f"num1: {num1}")
    print(f"num2: {num2}")
    print(f"num1이 num2보다 {num1 - num2}만큼 큽니다")
else:
    print("num1이 num2보다 크지 않네요")
```

```python
# python은 한 구문을 들여쓰기로 구분합니다
# 들여쓰기 조심!
if True:
    print("들여쓰기를")
     print("잘해야합니다")
```

```python
# elif를 통해 아닐 경우 + 조건을 추가할 수 있습니다
num1 = 10
num2 = 1
if num1 > num2:
    print("num1이 num2보다 큽니다")
elif num1 < num2:
    print("num1이 num2보다 작습니다")
else:
    print("num1과 num2는 같습니다")
```

```python
# 점수에 따라 성적을 출력해주는 프로그램
score = 65
if score >= 90:
    print('A')
elif score >= 80:
    print('B')
elif score >= 70:
    print('C')
elif score >= 60:
    print('D')
else:
    print('F')
```

```python
# 위의 코드 간결하게 만들어보기
# 이렇게도 생각할 수 있군..............

score = 65

# grades 만들고..
grades = ['F', 'F', 'F', 'F', 'F', 'F', 'D', 'C', 'B', 'A', 'A']

grades[score // 10]
```

### 여러 조건

```python
# 만약 고객이 성인이고, 남자라면 검정색을 추천해주세요
# 만약 고객이 여자이거나 초등학생이라면 핑크색을 추천해주세요
# 만약 고객의 나이가 15, 17세가 아니라면 파랑색을 추천해주세요
```

```python
# 여러 조건이 모두 만족해야할 때: and
# [조건1] and [조건2]
# 조건1도 참이고 조건2도 참이어야 참

client = {
    'gender': 'M',
    'age': 18
}
# 만약 고객이 성인이고, 남자라면 검정색을 추천해주세요

if client['gender'] == 'M' and client['age'] >= 20:
    print("검정색 추천한당")
else:
    print('추천 없음')
```

```python
# 여러 조건중 하나만 만족해도 될 때: or
# [조건1] or [조건2]
# 조건1이 참이거나 조건2가 참이면 참

client = {
    'gender': 'F',
    'age': 23
}
# 만약 고객이 여자이거나 초등학생이라면 핑크색을 추천해주세요
if client['gender'] == 'F' or client['age'] < 13:
    print("핑크색을 추천합니다")
```

```python
client = {
    'age': 16
}
# 만약 고객의 나이가 15, 17세가 아니라면 파랑색을 추천해주세요
if client['age'] not in [15, 17]:
    print("파랑색을 추천합니다")
if not client['age'] in [15, 17]:
    print("파랑색을 추천합니다")
```

```python
# 만약 고객이 남성이거나 초등학생 또는 고등학생이라면 빨강색을 추천해주세요
client = {
    'gender': 'F',
    'age': 18
}

if client['gender'] == 'M' or (client['age'] >= 8 and client['age']<= 13) or (client['age'] >= 17 and client['age'] <= 19):
    print("빨간색을 추천합니덩")
```

### 조건부 표현식

```python
gender = 'M'
message = ""
if gender == 'M':
    message = "남자"
else:
    message = "여자"

# 위 조건문을 심플하게 ... 조건부 표현식 (conditional expression)
message = "남자" if gender == 'M' else "여자"
print(message)

# 변수 = [참일 때 값] if [조건] else [거짓일 때 값]
```

```python
# 연습문제 1) 정수 x가 0 이상이면 "양수", 아니면 "음수"를 출력하라.
x = 1
msg = "양수" if x >= 0 else "음수"
print(msg)

# 정수 n이 짝수면 "짝수", 홀수면 "홀수"를 출력하라.
n = 2139
msg = "짝수" if n % 2 == 0 else "홀수"
print(msg)
```
