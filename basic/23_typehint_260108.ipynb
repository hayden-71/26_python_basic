## type hint

```python
# 함수 f는 어떤 함수일까요?
def f(a, b):
    return a + b

f(5, 3)
f("a", "b")
```

```python
# 변수 및 리턴 타입에 대하여 표기 가능
# f에 마우스 갖다대보슈

def f(a:int, b:int) -> int:
    """
    이것은 a와 b를 더해주는 함수입니당~
    """
    return a + b


name: str = "hello"
age: int = 20
```

```python
# 강제성은 없음!
def f(a:int, b:int) -> int:
    return a + b

name: int = 5
print(type(name))
f("c","d")
```
