### 클래스(Class)

```python
# 클래스: 특정 기능을 객체화 하여 만들기 위한 설계, 코드
# 데이터와 기능을 하나루 묶어 구조화

```

```python
# <기본 용어 설명>

# 클래스(class) : 객체를 만들기 위한 틀
# 객체(object) : 클래스라는 설계도로 만들어낸 실체
# 인스턴스(instance) : “특정 클래스에서 생성되었다”는 소속을 강조할 때 쓰는 말
# 속성(attribute) : 객체가 가지고 있는 속성 
# 메소드(method) : 객체가 수행할 수 있는 기능
```

```python
# <클래스 기본 형식>

# class 클래스이름:
#     def __init__(self, 인자1, 인자2, ...)
#         ~~~
#     def 메소드1:
#         ~~~
#     def 메소드2:
#         ~~~

# 클래스명 PascalCase 첫 문자 및 단어 시작 글자를 대문자로 표기 (ex: MyClass, Character)
```

```python
# <클래스의 self, __init__ 설명>

# self
# 파이썬 클래스에서 메서드를 정의할 때 첫 번째 인자 이름을 self로 쓰는 것이 관례
# 클래스에서 self의 의미는 인스턴스 '자기 자신'을 의미
# 클래스의 함수는 선언할 때 첫번째 인자는 self라고 선언하는 것이 원칙이고, 사용할 때는 self 빼고 사용

# init 메소드
# __init__은 생성자(constructor) 역할을 하는 특별 메서드
# 객체 선언 시 self는 제외하고, 정의한 매개변수(parameter)들은 객체를 만들 때 반드시 인자로 전달 필요
# __init__을 사용하면 클래스명을 쓰고 옆에 바로 인자들을 채워 넣음으로써 그 값들을 지닌 객체를 만들어 낼 수 있음
```

```python
class Bird:
    def __init__(self, name): # 클래스가 실제로 객체화될 때 동작하는 함수
        self.name = name
        print(f"{name} 생성")
    
    def cry(self):
        print(f"{self.name} {self.name}")

b1 = Bird("앵무새")
b2 = Bird("참새")

b1.cry()
b2.cry()
print(b1.name)
```

```python
class Charactor:
    def __init__(self, nickname, hp=100):
        self.nickname = nickname
        self.lv = 1
        self.hp = hp
        print(f"[{nickname}] 캐릭터 생성")

    def skill(self):
        print(f"[{self.nickname}]가 스킬을 사용한다")

    def level_up(self):
        self.lv += 1
        self.hp = self.lv * 100
        print(f"[{self.nickname}] Wow! Lv {self.lv}!!")

    def print_status(self):
        print(f"[{self.nickname}] Lv: {self.lv} / Hp: {self.hp}")


# damin 생성
damin = Charactor("damin")
damin.print_status()
damin.skill()
damin.level_up()
damin.level_up()


# hay 생성
hay = Charactor("hay")
damin.level_up()
damin.level_up()
damin.print_status()
hay.print_status()
```

```python
# [실습] 가위, 바위, 보 로봇 대전
# Robot은 game()을 통해 가위, 바위, 보 중 임의의 값을 제시
# 로봇 2개를 생성하여 가위바위보 대결
# 각 로봇은 승,무,패 숫자 기록
# 10번 후 각 로봇의 기록 출력
```

```python
import random
class Robot:
    def __init__(self):
        self.draw = 0
        self.win = 0
        self.lose = 0
        self.hand = ""

    def game(self):
        self.hand = random.choice(["R", "P", "S"])  # 여러개 고르는건 sample

    def record(self, other):
        if self.hand == other:
            self.draw += 1
        elif (
            (self.hand == "R" and other == "P")
            or (self.hand == "P" and other == "S")
            or (self.hand == "S" and other == "R")
        ):
            self.lose += 1
        else:
            self.win += 1

    def print_status(self):
        print(f"win: {self.win} draw: {self.draw} lose: {self.lose}")


r1 = Robot()
r2 = Robot()

for i in range(0, 10):
    r1.game()
    r2.game()
    r1.record(r2.hand)
    r2.record(r1.hand)

r1.print_status()
r2.print_status()
```

### 클래스 변수

```python
# 클래스 내 변수 선언 가능

class Family:
    lastname = "박" # 이렇게 하면 '클래스 자체'가 가지는 속성

daddy = Family()
son = Family()

print(f"family: {Family.lastname}")
print(f"daddy: {daddy.lastname}")
print(f"son: {son.lastname}")

Family.lastname = "김"

print(f"family: {Family.lastname}")
print(f"daddy: {daddy.lastname}")
print(f"son: {son.lastname}")

son.lastname = "최" # 별도로 선언..!
print(f"family: {Family.lastname}")
print(f"daddy: {daddy.lastname}")
print(f"son: {son.lastname}")
```

### 정적 메서드

```python
# 객체 생성하지 않고 클래스안에 포함된 일반 함수
# @staticmethod 로 함수 정의

class Calculator:
    @staticmethod # 들어온 값 정의해줌
    def add(a, b):
        return a + b

print(f" 1 + 2 = {Calculator.add(1, 2)}")
```

### 클래스 메서드

```python
# 클래스 변수에 접근이 필요할 경우
# @classmethod / 첫 인자로 cls 받음

class Calculator:
    operation = "-"
    @classmethod
    def op(cls, a, b):
        if cls.operation == "+":
            return a + b
        elif cls.operation == "-":
            return a - b

op = "-"
Calculator.operation = op
    
print(f"1 {op} 2 = {Calculator.op(1, 2)}")
```

### 상속 (Inheritance)

```python
# 상속(Inheritance)이란
# - 클래스를 부모와 자식으로 나눈 후 부모 클래스의 내용을 자식이 가져다 쓸 수 있는 것.
# - 상속을 해주는 클래스가 부모
# - 상속을 받는 클래스가 자식
```

```python
class Charactor:
    def __init__(self, nickname, hp):
        self.lv = 1
        self.nickname = nickname
        self.hp = hp
        print(f"[{nickname}] 캐릭터 생성")

    def skill(self):
        print(f"[{self.nickname}]가 스킬을 사용한다.")

    def level_up(self):
        self.lv += 1
        self.hp = self.lv * 100
        print(f"[{self.nickname}] Wow!! Lv {self.lv}!!")

    def print_status(self):
        print(f"[{self.nickname}] Lv: {self.lv} / HP: {self.hp}")
```

```python
class Knight(Charactor): # 부모: 캐릭터 클래스
    def __init__(self, nickname):
        super().__init__(nickname, 500)
        print(f"[{nickname}] 기사 탄생")

k1 = Knight("hay")
k1.skill
```

```python
# 오버라이딩(Overriding)이란
# 부모 클래스에게서 물려받은 메서드를 자식 클래스에서 재정의해서 사용하는 것
# 부모 메서드의 이름과 매개변수를 똑같이 정의하면, 자식 클래스의 메서드가 부모 메서드를 덮어쓴다.
# 즉, "부모의 기능은 유지하되, 자식에게 맞게 기능을 바꾸는 것"
```

```python
class Knight(Charactor):
    def __init__(self, nickname):
        super().__init__(nickname, 500)
        print(f"[{nickname}] 기사 탄생")

    def skill(self): # 다시 정의한 skill이 나옴
        print(f"[{self.nickname}] 칼을 휘둘휘둘")

    def ride(self):
        print(f"[{self.nickname}] 말을 타고 달립니다")


k1 = Knight("hay")
k1.skill()
k1.level_up()
k1.ride()
```

```python
# 다중 상속 - 여러개의 클래스 속성을 받을 수 있습니다
# 공통된 건 부모 클래스에.. 그리고 디테일한 것들은 새롭게 정의

class Unit:
    def __init__(self):
        self.hp = 100
    
    def move(self):
        print("앞으로")

class Monster:
    def __init__(self, size):
        self.size = size
    def attack(self):
        print("공격")

class Slime(Unit, Monster):
    pass

s1 = Slime()
s1.move()
s1.attack()
```
