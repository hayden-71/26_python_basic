### while문

```python
# 일정한 동작 반복 수행이 필요할 때 활용하는 반복문
# 조건이 만족(참)할 동안 계속 반복 수행
# while [조건]:
#   [수행]
```

```python
# a를 1씩 증가하며 출력하기: 5보다 작을 동안
a = 0

while a < 5:
    print(f"in: {a}")
    a += 1
print(f"out: {a}")
```

```python
# 입력 값 받기
prompt = """
1. Add / 2. Del / 3. List / 4. Quit
"""

number = 0

while number != 4:
    print(prompt)
    number = int(input()) # 사용자에게 값 받기
```

```python
# 피보나치 수열
fibo = [1, 1]
index = 0
while index < 10:
    fibo.append(fibo[index] + fibo[index + 1])
    index += 1
print(fibo)

fibo = [0, 1]
index = 0
while index < 10:
    fibo.append(fibo[index] + fibo[index + 1])
    index += 1
print(fibo)
```

```python
# 99까지 3, 6, 9 출력하기

number = 1
MAX = 99
while number <= MAX:
    if number % 10 in [3, 6, 9] or number // 10 in [3, 6, 9]: #1의 자리=나머지, 10의 자리=몫
        print("**", end=", ")
    else:
        print(f"{number:<2}", end=", ")
    if number % 10 == 0:
        print("") # 10개씩 줄바꿈
    number += 1


# 100의 자리까지 하려면? 숫자를 문자로 변환해서 ..?
print("")
number = 1
MAX = 999
while number <= MAX:
    if '3' in str(number) or '6' in str(number) or '9' in str(number):
        print("**", end=", ")
    else:
        print(str(number), end=", ")
    number += 1
```

```python
# 끝나지 않는 무한 loop
while True:
    print("hello")
```

```python
# 중간에 흐름 제어하기
# continue: 뒷부분 수행 않고 다시 위로
num = 0
while num < 10:
    num += 1
    if num % 3 == 0:
        continue # 3의 배수일 때는 뒷부분은 실행하지 않고, 조건문 위로 올라간다!
    print(f"in: {num}")


# 잘못 짠 예시.. num 값이 변하지 않게 되어서 무한루프..
# while num < 10:
#     if num % 3 == 0:
#         continue
#     print(f"in: {num}")
#     num += 1
```

```python
# 중간에 흐름 제어하기
# break: 해당 루프 탈출
num = 0

while True:
    if num == 5:
        break
    print(num)
    num += 1
```
